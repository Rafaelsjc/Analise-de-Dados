```python
#Importando bibliotecas
%matplotlib inline
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
plt.style.use('seaborn')
#Lendo o arquivo
filename = 'Algoritmo.csv'
df = pd.read_csv(filename, encoding = 'utf-8', error_bad_lines=False)
```

```python
df.index = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']
```

```python
df.drop('Total', axis = 1, inplace = True)
df.drop('Ano do Óbito', axis = 1, inplace = True)
df.drop('Ign', axis = 1, inplace = True)
```

```python
df
```

```python
df.plot.bar(title = 'TAXA DE SUICÍDIO POR SEXO')
plt.rcParams['figure.figsize'] = (20,10)
plt.xlabel('ANO')
```

```python
df_max_masc = df.nlargest(5, 'Masc')
df_max_masc.plot.bar(title = 'ANOS COM MAIORES TAXAS DE SUICÍDIO')
plt.xlabel('ANO')
```

```python
df_min = df.nsmallest(5, 'Masc')
df_min.plot.bar(title = 'ANOS COM MENORES TAXAS DE SUICÍDIO')
plt.xlabel('ANO')
```
