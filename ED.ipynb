```python
#Importando bibliotecas
%matplotlib inline
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
plt.style.use('seaborn')
#Lendo o arquivo
filename = 'ED.csv'
df = pd.read_csv(filename)
```

```python
df.rename(columns = {'year':'Ano'}, inplace = True)
#Mudando o nome dos índices
df.index = ['1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
```

```python
df_soma = df.sum(axis=1)
df_soma = pd.DataFrame(df_soma, columns=['Total de Desmatamentos'])
df_soma.plot.bar(title = 'TOTAL DE DESMATAMENTO POR ANO')
plt.xlabel('ANO')  
plt.ylabel('Desmatamento por KM²')
plt.rcParams['figure.figsize'] = (12,9)
```

```python
#Pegando os maiores de cada UF
df_maior_estado= [df.AC.max(), df.AM.max(), df.AP.max(), df.MA.max(), df.MT.max(), df.PA.max(), df.RO.max(), df.RR.max(), df.TO.max()]
df_maior_estado =  pd.DataFrame([df_maior_estado], columns=['AC', 'AM', 'AP', 'MA', 'MT', 'PA', 'RO', 'RR', 'TO'])

#Pegando os menores de cada UF
df_menor_estado = [df.AC.min(), df.AM.min(), df.AP.min(), df.MA.min(), df.MT.min(), df.PA.min(), df.RO.min(), df.RR.min(), df.TO.min()]
df_menor_estado =  pd.DataFrame([df_menor_estado], columns=['AC', 'AM', 'AP', 'MA', 'MT', 'PA', 'RO', 'RR', 'TO'])
```

```python
df_maior_estado.plot.bar(title = 'MAIORES DESMATAMENTOS POR UF')
plt.ylabel('Desmatamento por KM²')
df_maior_estado.index=['UF']
```

```python
df_menor_estado.plot.bar(title = 'MENORES DESMATAMENTOS POR UF')
plt.ylabel('Desmatamento por KM²')
df_menor_estado.index=['UF']
```
